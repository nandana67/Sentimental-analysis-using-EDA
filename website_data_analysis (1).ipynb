```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('/content/amazon.csv',encoding='unicode_escape')
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.drop(['img_link','product_link'],axis=1,inplace=True, errors='ignore')
```

```python
df['actual_price']=df['actual_price'].astype('int')
```

```python
df['actual_price'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'discounted_price':'discounted%'})
```

```python
df.describe()
```

```python
df[['product_name','rating','actual_price']].describe()
```

```python
ax=sns.countplot(x='rating',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['rating'],as_index=False)['actual_price'].sum().sort_values(by='actual_price',ascending=False)
sns.barplot(x='rating',y='actual_price',data=sales_gen)
```

```python
sales_gen=df.groupby(['discount_percentage'],as_index=False)['actual_price'].sum().sort_values(by='actual_price',ascending=False)
sns.barplot(x='discount_percentage',y='actual_price',data=sales_gen)
```

```python
sales_state = df.groupby(['product_id'], as_index=False)['actual_price'].sum().sort_values(by='actual_price', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'product_id',y= 'actual_price')
```
